```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_name = 'Construction_Data_PM_Tasks_All_Projects.csv'
data = pd.read_csv(file_name) 

# Convert the "Report Status" and "Task Group" columns to string
data['Report Status'] = data['Report Status'].astype(str)
data['Task Group'] = data['Task Group'].astype(str)

# Group by "Task Group" and "Report Status" and count the tasks
task_counts = data.groupby(['Task Group', 'Report Status']).size().unstack()

# Plot the data
task_counts.plot(kind='bar', stacked=False, figsize=(10,7))
plt.title('Total Number of Open and Closed Tasks by Each Task Group')
plt.ylabel('Number of Tasks')
plt.xlabel('Task Group')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_name = 'Construction_Data_PM_Tasks_All_Projects.csv'
data = pd.read_csv(file_name)

# Convert the "OverDue" and "Project" (assuming your column is named "Project", adjust if different) columns to string
data['OverDue'] = data['OverDue'].astype(str)
data['Project'] = data['Project'].astype(str)  # Adjust the column name if it's different

# Filter for overdue tasks and group by "Project" to count them
overdue_tasks = data[data['OverDue'].str.strip().str.lower() == 'true']
overdue_counts = overdue_tasks.groupby('Project').size()

# Plot the data
overdue_counts.plot(kind='bar', figsize=(10,7))
plt.title('Total Number of Overdue Tasks by Project')
plt.ylabel('Number of Overdue Tasks')
plt.xlabel('Project')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_name = 'Construction_Data_PM_Tasks_All_Projects.csv'
data = pd.read_csv(file_name) 

# Convert the "OverDue" and "project" columns to string
data['OverDue'] = data['OverDue'].astype(str)
data['project'] = data['project'].astype(str)

# Filter for overdue tasks and group by "project" to count them
overdue_tasks = data[data['OverDue'].str.strip().str.lower() == 'true']
overdue_counts = overdue_tasks.groupby('project').size()

# Plot the data
overdue_counts.plot(kind='bar', figsize=(10,7))
plt.title('Total Number of Overdue Tasks by Project')
plt.ylabel('Number of Overdue Tasks')
plt.xlabel('Project')
plt.tight_layout()
plt.show()
```
